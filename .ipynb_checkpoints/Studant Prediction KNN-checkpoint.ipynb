 # KNN
 ### Características:
  - Algoritmo simples e poderoso, mas preguiçoso;
  - Indicado quando o relacionamento entre as características é complexo;
  - Valor de K pequeno : dados com ruídos ou outliers podem prejudicar;
  - Valor de K grande : tendência a classificar a classe com mais elementos (overfitting), recomenda-se a utilizar valores padrões entre 3 ou 5, responsável por promover bons resultados;
  - Lento para fazer previsões;
  - Pode utilizar-se outras fórmulas além da distância euclidiana, como :
      - Coeficiente de Pearson
      - Índice de Tanimoto
      - City Block

```python
#Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

#Import a acurácia.
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
#Criei outra variavel para transformar alguns dados em categorias
df = pd.read_csv('StudentsPerformance.csv', dtype={'gender':'category','test preparation course':'category',
                                                  'race/ethnicity':'category', 'parental level of education':'category',
                                                   'lunch':'category'})
df.info()
```

```python
df['generos_grupos'] = df['gender'].cat.codes
df['etnicos_grupos'] = df['race/ethnicity'].cat.codes
df['parental_education_grupos'] = df['parental level of education'].cat.codes
df['luch_grupos'] = df['lunch'].cat.codes
df['test_preparation_grupos'] = df['test preparation course'].cat.codes
df.drop(columns=['gender','race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'], axis=1, inplace=True)
```

```python
#Criar uma clase previsores a partir da média aritmética entre as três categorias de prova
df['mean of tests']= df.loc[:,['math score','reading score','writing score']].mean(axis=1).round(1)
df.head()
```

```python
#Normalização:
def normaliza_df(df):
    for i in df.columns.drop('generos_grupos'):
        df[i] = (df[i] - df[i].mean())/df[i].std()
    return df
```

```python
df_norm = normaliza_df(df.copy(deep=True))
df_norm.head()
```

```python
#Separar as colunas preditoras da coluna predita
x_norm = df_norm.drop('generos_grupos', axis=1)
y = df['generos_grupos'].values
```

### Realiza a amostragem.

#### Entrada:
- x_norm : Dados preditores normalizados.
- y : dados predito.
- train_size : Percentual de dados que serão utilizados para induzir a hipótese.
- suffle : A amostragem é feita de forma aleatória.
- stratify : A amostragem é feita tentando manter o mesmo percentual de instâncias de cada classe (balanceando).

```python
#Amostragem estratificada
x_treino, x_teste, y_treino, y_teste = train_test_split(x_norm, y, train_size=0.8,
                                                       shuffle=True, stratify=y)
unique, counts = np.unique(y_treino, return_counts=True)
dict(zip(unique, counts))
```

```python
#Amostragem não estratificada
x_treino, x_teste, y_treino, y_teste = train_test_split(x_norm, y, train_size=0.8,
                                                       shuffle=True, stratify=None)
unique, counts = np.unique(y_treino, return_counts=True)
dict(zip(unique, counts))
```

```python
k = 5
knn = KNeighborsClassifier(n_neighbors=k, weights='uniform', metric='minkowski', p =2)
knn.fit(x_treino, y_treino)
```

```python
y_predit = knn.predict(x_teste)
y_predit_prob = knn.predict_proba(x_teste)
```

```python
#Computar a acurácia do 5-NN do Conjunto de Teste
acuracia_teste = accuracy_score(y_teste, y_predit)
print("Acurácia do 5-NN no Conjunto de Teste: {:.2f}".format(acuracia_teste))
```

```python
plot_confusion_matrix(knn,x_teste,y_teste)
plt.plot()
```
